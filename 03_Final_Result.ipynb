```python
# Perform feature scaling via standardization (z-score normalization):

metric = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'ysrelease']
categorial = ['Cis', 'D', 'Dis', 'E', 'F', 'Fis', 'G', 'Gis', 'A', 'Ais', 'B', '3/4', '4/4', '5/4','major']

std_scaler = StandardScaler()
data_features_nonscaled_metric = data_full.loc[:,metric]
data_features_categorial = data_full.loc[:,categorial]
#data_features_nonscaled = data_full.loc[:,['energy','valence','acousticness','speechiness','ysrelease']]
data_features = std_scaler.fit_transform(data_features_nonscaled_metric)
data_features = pd.DataFrame(data_features,columns=metric)
data_features = pd.concat([data_features, data_features_categorial],axis=1)
#data_full['mood_happy'] = np.where(data_full['Mood']=='happy',1,0)
target = data_full['mood']
```

```python
# Train and test split

X_train, X_test, y_train, y_test = train_test_split(data_features, target, test_size=0.3, random_state=24)
```

```python
# Train model and check accuracy

gbc_classifier = GradientBoostingClassifier(random_state=24)
gbc_classifier.fit(X_train, y_train)
accuracy = gbc_classifier.score(X_test, y_test)
print(accuracy)
```

```python
def mood_prediction_magic(track, artist):
  track_id = sp.search(q='artist:' + artist + ' track:' + track, type='track')
  track_id = track_id['tracks']['items'][0]['id']

  track_data = sp.audio_features(track_id)
  track_data = pd.DataFrame(track_data)
  #['ysrelease'] = 2022- data_full.release_year.astype('int32')
  track_data['major'] = track_data['mode']
  a = sp.track(track_id)
  meta_release_year = int(a['album']['release_date'][0:4])
  track_data['ysrelease'] = 2022- meta_release_year

  keys = ['C','Cis','D','Dis','E','F','Fis','G','Gis','A','Ais','B']
  df_keys = pd.get_dummies(track_data.key)
  df_keys.columns = keys
  df_keys.drop(columns='C', inplace=True)

  time_signatures = ['1/4','3/4', '4/4', '5/4']
  df_time_signature = pd.get_dummies(track_data.time_signature)
  df_time_signature.columns = time_signatures
  df_time_signature.drop(columns='1/4', inplace=True)

  track_data = track_data.drop(columns=['track_href', 'type','analysis_url', 'id', 'uri', 'mode','key','time_signature'])
  
  # combining data to one data set
  data_full = pd.concat([track_data, df_keys, df_time_signature], axis=1)

  features_for_training = data_full.iloc[:,:-1]
  features_for_training = features_for_training.reindex(data_features.columns, axis=1)
  track_data = track_data.reindex(features_for_training.columns, axis=1)

  std_scaler = StandardScaler()
  scale_fit = std_scaler.fit(features_for_training) #save the mean and std. dev computed for your data.
  data_features_input = scale_fit.transform(track_data) #use the above saved values to scale your single observation or batch observations.

  data_features_input = pd.DataFrame(data_features_input)
  data_features_input.columns = track_data.columns

  label = gbc_classifier.predict(data_features_input)
  return print(label)
```
